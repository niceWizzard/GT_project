```python
import pandas as pd
import folium
from data import locations_list
import osmnx as ox
from utils.location_graph import LocationGraph


results = {i.name: {} for i in locations_list}
```

```python
place_name = "Bulacan, Philippines"
locGraph = LocationGraph(place_name)
```

```python
from itertools import combinations
comb= combinations(locations_list, 2)

for i, (loc1, loc2) in enumerate(comb):
        print(f"Processing {i+1}")
        r1, d1 = locGraph.get_path(loc1, loc2)
        r2, d2 = locGraph.get_path(loc2, loc1)
        res = {
            "distance": max(d1, d2),
            "route": r1 if d1 > d2 else r2,
        }
        results[loc1.name][loc2.name] = res
        results[loc2.name][loc1.name] = res
```

```python
from itertools import combinations, permutations

# Check triangle inequality
for a_key, b_key, c_key in permutations(results.keys(), 3):
    ab = results[a_key][b_key]["distance"]
    bc = results[b_key][c_key]["distance"]
    ac = results[a_key][c_key]["distance"]
    s = ab + bc
    if s < ac - 1e-4 :
        print(f"BAD: {ab}+{bc} => {ab + bc} < {ac}")
        print(f"{a_key} to {b_key} > {c_key}")
```

```python
out_res = []
for a, b in combinations(results.keys(), 2):
    ab = results[a][b]["distance"]
    out_res.append((a,b, ab))
```

```python

pd.DataFrame(out_res, columns=["Loc1", "Loc2", "Distances(m)"]).to_csv("calculations/road_distances.csv")
```
