```python
import pandas as pd
import itertools

df = pd.read_csv("calculations/pass_distances.csv")

df
```

```python

locations =df["Loc1"].unique() 
results = []
# Iterate through all 2-location combinations (excluding origin)
for x, y, z in itertools.combinations(locations, 3):
    xy = df.query("Loc1 == @x and Loc2 == @y")["Distance (m)"].iloc[0]
    yz = df.query("Loc1 == @y and Loc2 == @z")["Distance (m)"].iloc[0]
    xz = df.query("Loc1 == @x and Loc2 == @z")["Distance (m)"].iloc[0]

    valid = (xy + yz >= xz) and (xz + yz >= xy) and (xy + xz >= yz)

    results.append({
        "Triplet": (x, y, z),
        "Sides (m)": (round(xy, 2), round(yz, 2), f"{round(xz, 2)}"),
        "Valid Triangle": valid
    })

out_df = pd.DataFrame(results)
out_df
```
